```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
%matplotlib inline
```

```python
dataFrame=pd.read_csv("C:\\Users\\Sena\\Desktop\\RA-PYTHON\\PROJECT1\\Cost_of_Living_Index_by_Country_2024.csv")
```

```python
plt.figure(figsize=(14, 10))
dataFrame[['Cost of Living Index', 'Rent Index', 'Cost of Living Plus Rent Index', 'Groceries Index',
           'Restaurant Price Index', 'Local Purchasing Power Index']].hist(bins=30, figsize=(14, 10), layout=(3, 2))
plt.tight_layout()
```

```python
plt.figure(figsize=(20, 8))
for column in ["Cost of Living Index", "Rent Index", "Cost of Living Plus Rent Index","Groceries Index","Restaurant Price Index",
               "Local Purchasing Power Index"]:
    plt.plot(dataFrame["Country"], dataFrame[column], label=column)
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Index Value")
plt.title("Distribution of Indices by Country")
plt.legend()
```

```python
fig, axes = plt.subplots(nrows=6, ncols=1, figsize=(18, 30)) 

sbn.barplot(x="Country", y="Cost of Living Index", data=dataFrame, ax=axes[0],color="#00ACBE")
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=90)
axes[0].set_title("COST OF LIVING INDEX BY COUNTRY")
axes[0].set_xlabel("Country")
axes[0].set_ylabel("Cost of Living Index")

sbn.barplot(x="Country", y="Rent Index", data=dataFrame, ax=axes[1],color="#D70063")
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=90)
axes[1].set_title("RENTAL INDEX BY COUNTRY")
axes[1].set_xlabel("Country")
axes[1].set_ylabel("Rent Index")

sbn.barplot(x="Country", y="Cost of Living Plus Rent Index", data=dataFrame, ax=axes[2])
axes[2].set_xticklabels(axes[2].get_xticklabels(), rotation=90)
axes[2].set_title("COST OF LIVING PLUS RENT INDEX BY COUNTRY") 
axes[2].set_xlabel("Country")
axes[2].set_ylabel("Cost of Living Plus Rent Index")

sbn.barplot(x="Country", y="Groceries Index", data=dataFrame, ax=axes[3],color="#00ACBE")
axes[3].set_xticklabels(axes[3].get_xticklabels(), rotation=90)
axes[3].set_title("GROCERY INDEX BY COUNTRY") 
axes[3].set_xlabel("Country")
axes[3].set_ylabel("Groceries Index")

sbn.barplot(x="Country", y="Restaurant Price Index", data=dataFrame, ax=axes[4],color="#D70063")
axes[4].set_xticklabels(axes[4].get_xticklabels(), rotation=90)
axes[4].set_title("RESTAURANT PRICE INDEX BY COUNTRY")
axes[4].set_xlabel("Country")
axes[4].set_ylabel("Restaurant Price Index")

sbn.barplot(x="Country", y="Local Purchasing Power Index", data=dataFrame, ax=axes[5])
axes[5].set_xticklabels(axes[5].get_xticklabels(), rotation=90)
axes[5].set_title("LOCAL PURCHASING POWER INDEX BY COUNTRY") 
axes[5].set_xlabel("Country")
axes[5].set_ylabel("Local Purchasing Power Index")

plt.tight_layout() 
```

```python
fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(18, 30))  

sbn.barplot(x="Cost of Living Index", y="Rent Index", data=dataFrame, ax=axes[0],color="#D70063")
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=90)
axes[0].set_title("COST OF LIVING INDEX BASED ON RENTAL INDEX")
axes[0].set_xlabel("Cost of Living Index")
axes[0].set_ylabel("Rent Index")

sbn.barplot(x="Cost of Living Index", y="Cost of Living Plus Rent Index", data=dataFrame, ax=axes[1])
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=90)
axes[1].set_title("COST OF LIVING PLUS RENT INDEX ACCORDING TO COST OF LIVING INDEX")
axes[1].set_xlabel("Cost of Living Index")
axes[1].set_ylabel("Cost of Living Plus Rent Index")

sbn.barplot(x="Cost of Living Index", y="Groceries Index", data=dataFrame, ax=axes[2],color="#00ACBE")
axes[2].set_xticklabels(axes[2].get_xticklabels(), rotation=90)
axes[2].set_title("GROCERY INDEX BY COST OF LIVING INDEX") 
axes[2].set_xlabel("Cost of Living Index")
axes[2].set_ylabel("Groceries Index")

sbn.barplot(x="Cost of Living Index", y="Restaurant Price Index", data=dataFrame, ax=axes[3],color="#D70063")
axes[3].set_xticklabels(axes[3].get_xticklabels(), rotation=90)
axes[3].set_title("RESTURANT PRİCE INDEX BY COST OF LIVING INDEX") 
axes[3].set_xlabel("Cost of Living Index")
axes[3].set_ylabel("Restaurant Price Index")

sbn.barplot(x="Cost of Living Index", y="Local Purchasing Power Index", data=dataFrame, ax=axes[4])
axes[4].set_xticklabels(axes[4].get_xticklabels(), rotation=90)
axes[4].set_title("LOCAL PURCHASING POWER INDEX BY COST OF LIVING INDEX") 
axes[4].set_xlabel("Cost of Living Index")
axes[4].set_ylabel("Local Purchasing Power Index")

plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(18, 20))  

sbn.barplot(x="Rent Index", y="Cost of Living Plus Rent Index", data=dataFrame, ax=axes[0])
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=90)
axes[0].set_title("RENTAL INDEX PLUS COST OF LIVING ACCORDING TO RENTAL INDEX")
axes[0].set_xlabel("Rent Index")
axes[0].set_ylabel("Cost of Living Plus Rent Index")

sbn.barplot(x="Rent Index", y="Groceries Index", data=dataFrame, ax=axes[1],color="#00ACBE")
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=90)
axes[1].set_title("GROCERY INDEX ACCORDING TO RENT INDEX") 
axes[1].set_xlabel("Rent Index")
axes[1].set_ylabel("Groceries Index")

sbn.barplot(x="Rent Index", y="Restaurant Price Index", data=dataFrame, ax=axes[2],color="#D70063")
axes[2].set_xticklabels(axes[2].get_xticklabels(), rotation=90)
axes[2].set_title("RESTAURANT PRICE INDEX ACCORDING TO RENTAL INDEX") 
axes[2].set_xlabel("Rent Index")
axes[2].set_ylabel("Restaurant Price Index")

sbn.barplot(x="Rent Index", y="Local Purchasing Power Index", data=dataFrame, ax=axes[3])
axes[3].set_xticklabels(axes[3].get_xticklabels(), rotation=90)
axes[3].set_title("LOCAL PURCHASING POWER INDEX ACCORDING TO RENT INDEX") 
axes[3].set_xlabel("Rent Index")
axes[3].set_ylabel("Local Purchasing Power Index")

plt.tight_layout()
```
